```python
def fib(num: int) -> int:
    if num < 2:
        return num
    else:
        return fib(num - 1) + fib(num - 2)


def imprimirValor(valor: int):
    print(valor)


def main():
    n = 9
    resultado = fib(n)
    imprimirValor(resultado)


if __name__ == "__main__":
    main()
```

```python
def factorial(x: int) -> int:
    if x == 0:
        return 1
    else:
        return x * factorial(x - 1)


def validarYCalcular(valor: int):
    if valor < 0:
        print("Error: factorial no definido")
    else:
        r = factorial(valor)
        print("Resultado factorial =", r)


def main():
    n = 5
    validarYCalcular(n)


if __name__ == "__main__":
    main()
```

```python
def esPar(x: int) -> bool:
    return x % 2 == 0


def sumaHasta(n: int) -> int:
    total = 0
    while n > 0:
        total += n
        n -= 1
    return total


def main():
    numero = 7

    # Estructura condicional
    if esPar(numero):
        print(numero, "es par")
    else:
        print(numero, "es impar")

    # Ciclo cuenta regresiva
    i = numero
    while i > 0:
        print(i)
        i -= 1

    # Suma 1..n
    print("Suma desde 1 hasta", numero, "=", sumaHasta(numero))


if __name__ == "__main__":
    main()
```
